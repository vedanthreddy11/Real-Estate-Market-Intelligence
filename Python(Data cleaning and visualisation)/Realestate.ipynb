**CATEGORICAL DATA SHOULD BE FILLED WITH MODE AND NUMERICAL COLUMNS WITH MEDIAN**

```python
df.Location.astype(object)
```

```python
df.Bedrooms.astype(float).mean()
```

```python
df['Location'].fillna(df['Location'].mode()[0],inplace=True)
df['Property_Type'].fillna(df['Property_Type'].mode()[0],inplace=True)
df['Condition'].fillna(df['Condition'].mode()[0],inplace=True)df['Condition'].fillna(df['Condition'].mode()[0],inplace=True)
df['Bedrooms'].fillna(df['Bedrooms'].median(),inplace=True)
df['Bathrooms'].fillna(df['Bathrooms'].median(),inplace=True)
df['Square_Feet'].fillna(df['Square_Feet'].median(),inplace=True)
df['Year_Built'].fillna(df['Year_Built'].median(),inplace=True)
df['Price'].fillna(df['Price'].median(), inplace=True)

```

**NOW THE DATA IS FILLED WITH VALUES COMPLETLY**

*DATA VISUALISATION USING MATPLOTLIB*

**1.DISTRIBUTION OF PROPERTY PRICES**

```python
plt.figure(figsize=(8,5))
plt.hist(df['Price'].dropna(), bins=30, edgecolor='black')
plt.title("Distribution of Property Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

**2.HOW MANY PROPERTIES ARE THERE IN EACH LOACTION**

```python
plt.figure(figsize=(8,5))
df['Location'].value_counts().plot(kind='bar', color='orange', edgecolor='black')
plt.title("Number of Properties by Location")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

**3.REALTION BETWEEN SQUARE FEET AND PRICE**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Square_Feet'], df['Price'], alpha=0.6, c='green')
plt.title("Square Feet vs Price")
plt.xlabel("Square Feet")
plt.ylabel("Price")
plt.show()
```

**4.BEDROOMS AFFECT THE PRICE**

```python
plt.figure(figsize=(8,5))
df.groupby('Bedrooms')['Price'].mean().plot(marker='o', linestyle='-', color='purple')
plt.title("Average Price vs Number of Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Average Price")
plt.show()
```

**5.PRICE OF PROPERTY BASED ON PROPERTY TYPE**

```python
plt.figure(figsize=(8,5))
avg_price = df.groupby('Property_Type')['Price'].mean()
avg_price.plot(kind='bar', color='yellow', edgecolor='black')
plt.title("Average Property Price by Type")
plt.ylabel("Average Price")
plt.show()
```

                                         ***OVERALL INSIGHTS***

The analysis shows that most properties fall within the lower-to-mid price range, with luxury homes being relatively rare. 
Larger homes and those with more bedrooms generally command higher prices, though location and condition often cause significant variations.
Townhouses tend to have the highest average prices compared to condos and apartments.Overall,
property size and type strongly influence value, but location remains the most critical factor in pricing.
